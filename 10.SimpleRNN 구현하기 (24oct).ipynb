```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense
import numpy as np
```

```python
## 문제데이터
X = np.array([   ##(하나의)샘플을 감싸는 대괄호[]
    [
        [1,0,0,0],  # h
        [0,1,0,0],  # e
        [0,0,1,0],  # l
        [0,0,1,0],  # l
    ]
])

X.shape   ## 학습시킬샘플수(hello 한개!),  
          ## 4개의timestep(hell 다음을 맞추기위한 4개의 timestep),  
          ## 4개의 features(h e l o)
```

```python
## 정답데이터
y = np.array([[0,0,0,1]])     # o
y.shape
```

### 모델링

```python
model = Sequential()

model.add(SimpleRNN(5, input_shape=(4,4)))   ##순환하는 RNN 뉴런이 5개가 붙어있다는것, 
                              ## 학습시키려는 데이터의 모양 = (4개의 time Step, time Step은 4개의 feature로 구성되어 들어간다~~)

model.add(Dense(10, activation='relu'))   ## 일반신경망(은닉층?)

model.add(Dense(4, activation='softmax'))  ## 정답이 0,0,0,1로 구성되어 있기때문에 출력층에 4   
                                           ## 나오는숫자값(숫자합)을 1로 만들어주는 softmax
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="Adam")
```

```python
model.fit(X, y, epochs=2000)
```

```python
model.predict(X)
## 3.6899579e-05 는 소수점 5자리 앞으로
## 5.5977656e-04 는 4자리 앞으로...
## 9.9937016e-01 는 한자리만 앞으로 ==> 1에 가깝게 출력
```

## 로이터 뉴스 분류하기
- 텍스트마이닝은 토큰화,수치화 필요
    - 토큰화

        - 띄어쓰기 단위  (나는,오늘,집에,간다)
        - N-gram   (1개씩? -> 나는,오늘,집에,간다  // 2개씩? -> 나는 오늘, 오늘 집에, 집에 간다 // 3개씩 --> 나는 오늘 집에, 오늘 집에 간다) 많이 묶지는 않음. 일반적으로 1~3사이에서 사용

### 모델링

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, Dropout
```

```python
model = Sequential()

model.add(SimpleRNN(32, input_shape=(100,1)))   ## 순환하는 RNN 몇개를 붙일것인가 timeStep100, 한번돌아갈때 feature한개씩 들어감
                                               ### 100개의 단어를 보고 이게 어떤 특성(1개)을 가지고 있는것인지 예측하는거지? 그렇기때문에 input_shape=(100,1) 
model.add(Dense(64, activation='relu'))

model.add(Dense(46, activation='softmax'))      ## 46개로 나오니 46개, 그리고 총합이 1로나와야하므로 softmax
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
## 학습하고 시각화시키기위해 history에 담아줌
history = model.fit(X_train, y_train_one_hot,
                   validation_split=0.3,   ##검증셋도 설정. 30퍼센트로 설정해줌
                   epochs=20)

## cf. fit() 한 번 돌린상태에서(eg.,epochs=20일때 20번돌아감) 다시 한 번 돌리면 40번 돌린 것이다
```

```python
plt.figure(figsize=(15,10))
plt.plot(history.history['loss'], label='loss')           ## 예측과 실제의 차이 loss. 이게 떨어져야 함~~
plt.plot(history.history['val_loss'], label='val_loss')   ## 검증데이터의 loss값 보면서 상승하지 않는지 살펴보며 과대적합인지 알아낼 수 있음
plt.legend()
plt.show()
```

### 장단기 기억메모리 RNN (LSTM 사용하기)
- 기존에 있는것만 간단하게 넘겨준 SimpleRNN 
    - 문제점: 길게 늘어질 경우(timeStep이 길어질수록) 처음에 넣었던 데이터가 희미해짐(변형).. 성능이 좀 안 좋음

- LSTM(Long SHort Term Memory)
    - 기억을 관리해주는 장치를 추가해놓음 <어떤게 더 유지/삭제 되어야하는 기억인지 관리>
    - LSTM이 복잡해서 이를 단순하게 만들어 놓은 GRU

```python
from tensorflow.keras.layers import LSTM
```

```python
model2 = Sequential()

model2.add(LSTM(32, input_shape=(100,1), return_sequences=True))
model2.add(Dropout(0.2))  ##층마다 활성화되는 LSTM 몇개를 막아줄 수 있음
model2.add(LSTM(32, return_sequences=True))   ### 위에 return_sequences=True 써줘야 층을 더 쌓아 줄 수 있음
                            ## 왜?? ====> 이거 써주면 <<"매"순환마다(=매회차마다)>> 결과를 도출하도록 바꿔준 것임 
                                         ##### (안썼을 경우 최종순환끝났을때만 연산결과를 뽑아낼 수 있음)
                            ## 나온 결과물을 다음 층에 넘겨주는 것임
                            ########### 쌓으려는 층 앞쪽에 return_sequences=True 써주면 됨
model2.add(Dropout(0.2))
model2.add(LSTM(32))


model2.add(Dense(64, activation='relu'))

model2.add(Dense(46, activation='softmax'))
```

```python
model2.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
history2 = model2.fit(X_train, y_train_one_hot,
          validation_split=0.3,
          epochs=20)
```

```python
plt.figure(figsize=(15,10))
plt.plot(history2.history['loss'], label='loss')           
plt.plot(history2.history['val_loss'], label='val_loss')   
plt.legend()
plt.show()
```

- cf. 결과물은 하나지만 여러번데이터를 넣는것과 같은 효과를 낼 수 있는 RNN의 RepeatVector 
    - (return_sequences=True를 사용하지 않고 층을 더 쌓을 수 있음 ----> 번역할 데이터 만들때?와 같은 구조)
    
    

- 성능 올리고 싶다면 층 더쌓아도 되지만 "수치화"를 더 잘해준다면 좀 더 성능이 좋아질 것이다
- eg., 단순 단어 빈도수가 아닌, 단어와의 "연관성"을 고려하게 만들어 보기!
