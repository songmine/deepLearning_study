#### AND 논리를 이용한 신경망

```python
import pandas as pd
import numpy as np

# AND 논리 데이터
XData = [[0, 0], [0, 1], [1, 0], [1, 1]]
y = [0, 0, 0, 1]
```

```python
# w와 b를 초기화 (항상 같은 값으로 w,b가 초기화되도록 함)
seed = 7                        ##seed 숫자는 자기 맘
np.random.seed(seed)
tensorflow.random.set_seed(seed)
```

#### 신경망 설계

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Sequential : 신경망을 구성해주는 기능
# Dense      : 신경망 층을 설정해주는 기능
# Activation : 활성화 함수를 설정해주는 기능
```

```python
model = Sequential() ##신경망 초기화

# 입력층
# add() : 신경망에 신경망 층을 추가
# 노드의 출력수, 노드의 입력수(특성수)
# input_dim : 입력층의 개수('특성'수)
model.add(Dense(1, input_dim=2))  ##임의의 수
#model.add(Activation("relu"))     ##입력층에서 활성화함수는 relu쓰면됨..ㅎ

# 출력층
# 회귀 : linear (|| 안써도 됨)
# 이진분류 : sigmoid, softmax
# 다중분류 : softmax
#model.add(Dense(1))
model.add(Activation("sigmoid"))

# summary() : 신경망 구조를 표시
model.summary()

##2개의 특성을 17개의 파라미터로 늘려놨음~~
```

```python
# compile() : 학습이 가능한 모델로 변환
# loss : 손실함수
#    - 회귀형 손실함수 : mse(mean squared error)
#    - 이진분류.. : binary_crossentropy            ##0이냐 1이냐
#    - 다중분류.. : categorical_crossentropy
# optimizer : 최적화함수 (of 경사하강법) - adam이 제일 나음
# metrics : 평가방법 : acc(accuracy)
model.compile(loss="binary_crossentropy", 
              optimizer="adam", metrics=["acc"]) 
```

```python
# 학습
# 특성데이터(X), 라벨데이터(y)
# epochs : w, b를 몇 번 갱신할 것인지 (= 몇 번 반복할 것인지)
# batch_size : 한 번 반복할 때 몇 개의 데이터를 이용할 것인지
model.fit(XData, y, epochs=1500, batch_size=4)
```

#### 예측

### 직접 논리 퍼셉트론(perceptron)을 구현해보자!

AND 논리

```python
# 입력 파라미터가 2개이고 출력파라미터(반환값)가 1개인 함수
def AND_Perceptron(x1, x2): #입력으로2개받고 출력(return)으로 1개반환하는 퍼센트론의 기본적인 구조!
    # 파라미터 초기화
    w1, w2, b = 0.5, 0.5, -0.5
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5:
        return 1
    else:
        return 0

print(AND_Perceptron(0,0))
print(AND_Perceptron(0,1))
print(AND_Perceptron(1,0))
print(AND_Perceptron(1,1))
```

OR 논리

```python
def OR_Perceptron(x1, x2):
    # 파라미터 초기화
    w1, w2, b = 0.5, 0.5, 0.3
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5:
        return 1
    else:
        return 0

print(OR_Perceptron(0,0))
print(OR_Perceptron(0,1))
print(OR_Perceptron(1,0))
print(OR_Perceptron(1,1))
```

NAND 논리

```python
# NAND -> NOT AND
def NAND_Perceptron(x1, x2):
    # 파라미터 초기화
    w1, w2, b = -0.1, -0.1, 0.7
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5:
        return 1
    else:
        return 0

print(NAND_Perceptron(0,0))
print(NAND_Perceptron(0,1))
print(NAND_Perceptron(1,0))
print(NAND_Perceptron(1,1))
```

XOR 논리 (exclusive OR)


```python
def XOR_Perceptron(x1, x2):
    # 파라미터 초기화
    w1, w2, b = 0.5, -0.3, 0.1
    
    temp = w1 * x1 + w2 * x2 + b
    
    if temp >= 0.5:
        return 1
    else:
        return 0

print(XOR_Perceptron(0,0))
print(XOR_Perceptron(0,1))
print(XOR_Perceptron(1,0))
print(XOR_Perceptron(1,1))
```

```python
def XOR_Perceptron(x1, x2):
    s1 = NAND_Perceptron(x1, x2)
    s2 = OR_Perceptron(x1, x2)
    y = AND_Perceptron(s1, s2)
    
    return y
    
print(XOR_Perceptron(0,0))
print(XOR_Perceptron(0,1))
print(XOR_Perceptron(1,0))
print(XOR_Perceptron(1,1))
```

#### 오차역전파 구현
- : 오차를 점점 거슬러 올라가면서 다시 전파하는 것
- 1개의 노드로 구성된 1층 신경망


```python
import numpy as np

seed = 0 #항상같은값나오게(초기화)하기위한 시드
np.random.seed(seed)

# 특징값, 반복수, 학습률
def backpropagation1(X, epochs, lr):
    # (step1) w, b를 초기화
    w = np.random.rand(1)      # 0에서 1사이 값 중에서 한 개를 추출
    b = np.random.rand(1)
    
    # (step2) epochs만큼 반복해서 학습
    for i in range(epochs):
        # (step2-1) 첫 번째 반복에서 X를 입력해서 y를 구한다
        pred_y = w*X + b 
        
        # (step2-2) 오차를 계산
        error = X - pred_y
        
        print("반복수 : {}, 예측값 : {}, 오차 : {}".format(i, pred_y, error))
        
        # (step2-3) 오차 역전파
        # 기울기 : 오차 / 예측값 (='미분')
        d1 = error / pred_y
        
        # w와 b를 갱신
        # lr : 학습률 (기울기를 w나 b를 갱신하는데 얼마만큼 비율을 적용할 것인지)
        w = w + lr * d1
        b = b + lr * d1
        
backpropagation1(10, 100, 0.01)
```

- 2개의 노드로 구성된 2개의 신경망


```python
import numpy as np

np.random.seed(0)

def backpropagation2(X, epochs, lr):
    # (1) w, b를 초기화
    w1 = np.random.rand(1)
    b1 = np.random.rand(1)
    w2 = np.random.rand(1)
    b2 = np.random.rand(1)
    
    for i in range(epochs):
        # (2) 층별 예측값 계산
        pred_h = w1*X + b1
        pred_y = w2*pred_h + b2
        
        # (3) 2층 오차 계산
        error2 = X - pred_y
        
        print("반복수 : {}, 예측값 : {}, 오차 : {}".format(i, pred_y, error2))
        
        # (4) 2층 오차 역전파(=미분)
        # 에러 / 예측값 (==미분)
        d2 = error2 / pred_y
        
        # (5) w2, b2 갱신
        w2 = w2 + lr * d2
        b2 = b2 + lr * d2
        
        # (6) 1층 오차 역전파
        d1 = pred_y / pred_h * d2
        
        # (7) w1, b1 갱신
        w1 = w1 + lr * d1
        b1 = b1 + lr * d1
        
backpropagation2(10, 100, 0.01)
```
