### 1. 데이터 가져오기 (로또번호, 날씨)

```python
import pandas as pd

lotto = pd.read_csv("./Data/lotto_869.csv", encoding="euc-kr")
lotto.head()
```

```python
# 추첨일을 인덱스 설정
lotto = lotto.set_index("추첨일")
```

```python
## 전처리

# 필요한 컬럼만 추출 (회차, 추첨번호 6개)
lotto = lotto.loc[:,["회차","No_1","No_2","No_3","No_4","No_5","No_6","No_Bonus"]]   ##전체rows,필요컬럼들
lotto.head()
```

```python
# 날씨 데이터 가져오기
weather = pd.read_csv("./Data/weater_seoul_200212_20190727.csv", encoding="EUC-KR")

weather = weather.set_index("일시")
weather.head()
```

```python
# 로또 데이터와 날씨 데이터를 병합
lotto_weather = pd.concat([lotto, weather], axis=1, join="inner")   #outer:합집합(union), inner:교집합(intersection)
lotto_weather.index.names=["일시"]
lotto_weather.head()
```

```python
lotto_weather.shape
```

```python
lotto_weather.drop(["회차","지점","최저기온","최고기온","평균 풍속","합계 일조 시간","평균 전운량","평균 지면온도"],
                  axis="columns", inplace=True)
lotto_weather.head()
```

##### 라벨을 추가하여 1로 입력 (1등 데이터)

```python
lotto_weather["등수"] = 1
lotto_weather.head()
```

###### 2등 데이터를 추가하고 라벨 설정

```python
 lotto_weather.iloc[:, 0]
```

```python
# 데이터 복사
lotto_weather2 = lotto_weather.copy()

# 6개 당첨번호를 각각 보너스 번호 변경
# 869개의 데이터에 대해 반복
for i in range(len(lotto_weather)):
    # 각 회차 데이터의 당첨번호에 대해 반복 (보너스 번호로 변경)
    for j in range(6):
         # 869개 데이터에서 1회차 데이터만 불러온다
        lotto_weather3 = lotto_weather.iloc[[i],:]      ##각각의 날짜에 해당하는 정보 가져옴(왜? 그 날짜의 2등번호들 더해줄것이기 때문)
                                                        ##현재는 1등한 사람의 번호만 들어가 있음

        lotto_weather3.iloc[:, j] = lotto_weather3.loc[:, "No_Bonus"]  #모든행의 No_Bonus열    
                                                                       ###--> 1등숫자 6개 중 한개씩에 바꿔치게 되게 해줌
        # 등수를 2등을 저장
        lotto_weather3.loc[:, "등수"] = 2
        
        # 새로운 당첨번호 2등 데이터 행을 추가
        lotto_weather2 = pd.concat([lotto_weather2, lotto_weather3])
```

```python
lotto_weather2.drop(["No_Bonus"], axis="columns", inplace=True)

lotto_weather2.head()
```

#### 특성데이터와 라벨데이터로 분리

```python
X = lotto_weather2.iloc[:, :-1]
y = lotto_weather2.iloc[:, -1]

X.shape, y.shape
```

#### 훈련데이터와 테스트데이터로 분리

```python
# One-hot Encoding
y_en = pd.get_dummies(y)    ##현 데이터는 2진분류(1등or2등)라 그냥해도되겠지만, 1~5(등)까지 해줘야함
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_en, random_state=777)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### 딥러닝

```python
import numpy as np
import tensorflow
seed = 0 
np.random.seed(seed)
tensorflow.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# 입력층
model.add(Dense(32, input_dim=X_train.shape[1], activation="relu"))  #X_train.shape[1] = 9

# 은닉층    
model.add(Dense(64, activation="relu"))
model.add(Dense(32, activation="relu"))
model.add(Dense(16, activation="relu"))
#model.add(Dense(8, activation="relu"))

# 출력층
model.add(Dense(2, activation="softmax"))

#model.summary()
```

```python
model.compile(loss="categorical_crossentropy",
             optimizer="adam",
             metrics=["acc"])
```

```python
history = model.fit(X_train, y_train, epochs=100 ,batch_size=50)
```

```python
model.evaluate(X_test,y_test)
```
